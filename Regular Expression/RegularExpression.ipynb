Finding the Bithplace

```python
pattern = 'Born\s[a-zA-z ]*\n[a-zA-z0-9 ,()]*\n([a-zA-z, ]*)'
matches = re.findall(pattern,text)
matches
```

```python
def pattern_match(pattern,text):
    matches = re.findall(pattern,text)
    if matches:
        return matches[0]
```

```python
pattern_match('Born\s[a-zA-z ]*\n[a-zA-z0-9 ,()]*\n([a-zA-z, ]*)',text)
```

```python
def get_personal_info(text):
    age = pattern_match('age.(\d+)',text)
    name = pattern_match('Born\s([a-zA-z ]*)\n',text)
    birthdate = pattern_match('Born\s[a-zA-z ]*\n([a-zA-z0-9 ,]*)',text)
    birthplace = pattern_match('Born\s[a-zA-z ]*\n[a-zA-z0-9 ,()]*\n([a-zA-z, ]*)',text)
    return{
        'Name': name.strip(),
        'Age' : int(age),
        'Birth Date':birthdate.strip(),
        'Birth Place':birthplace.strip()}
    
```

```python
get_personal_info(text)
```
