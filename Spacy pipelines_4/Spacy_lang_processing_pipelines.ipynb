```python
import spacy

nlp = spacy.blank("en")

doc = nlp("Captain america ate 100$ of samosa. Then he said I can do this all day.")

for token in doc:
    print(token)
```

```python
nlp.pipe_names
```

Pretrained pipeline in spacy

```python
nlp = spacy.load('en_core_web_sm')
nlp.pipe_names
```

```python
doc = nlp("Captain america ate 100$ of samosa. Then he said I can do this all day.")

for token in doc:
    print(token, "|", token.pos_, "|", token.lemma_)
```

```python
doc = nlp("Tesla Inc is going to acquire twitter for $45 billion")
for ent in doc.ents:
    print(ent, "|", ent.label_, "|", spacy.explain(ent.label_))
```

```python
from spacy import displacy
displacy.render(doc,style='ent')
```

Adding components to blank spacy from en_core_web_sm spacy

```python
source_nlp = spacy.load('en_core_web_sm')

blank_pipe = spacy.blank('en')

blank_pipe.add_pipe('ner', source=source_nlp)
```

```python
blank_pipe.pipe_names
```

```python
doc = blank_pipe("Tesla Inc is going to acquire twitter for $45 billion")
for ent in doc.ents:
    print(ent.text, ent.label_)
```
